# 8.1 Consultas em SQL

### **Qual é o valor total dos salários pagos, baseado no título da posição ocupada pelos funcionários?**

Para responder essa pergunta, usamos a instrução "SELECT" para recuperar os dados da tabela "salary" e realizar a operação de agregração, com o intuito de calcular a soma dos salários agrupados por "position\_title".

```python
SELECT position_title, 
SUM(salary) as total
FROM salary
GROUP BY position_title
```

### **Quais são os cargos que mais ocorrem nos dados?**

Para descobrir quais são os títulos de cargos que mais aparecem nos dados, selecionou-se a coluna "position\_title" e, usando a função "COUNT", contou-se quantas vezes cada título aparece na tabela salary. A cláusula "GROUP BY" foi declarada para agrupar os dados pela coluna "position\_title" e, a cláusula "ORDER BY" e "DESC" foram utilizdas para classsificar os resultados em ordem decrescente (do maior para o menor) com base no número de ocorrências (total\_ocorrencias).

```python
SELECT position_title, COUNT (*) as total_ocorrencias 
FROM salary
GROUP BY position_title
ORDER BY total_ocorrencias DESC
```

### **Quais são os 10 cargos com maiores salários?**

 A consulta tem como objetivo retornar as 10 primeiras linhas dessa tabela, ordenadas pelo campo "salary" em ordem decrescente (do maior para o menor salário). Dessa forma, usou-se "SELECT TOP 10 position\_title, salary" para selecionar apenas as 10 primeiras linhas da tabela "salary", retornando apenas as colunas "position\_title" e "salary". A cláusula "ORDER BY salary DESC" foi utilizada para ordenar os dados de forma decrescente, com base na coluna "salary"

```python
SELECT TOP 10 position_title, salary
FROM salary
ORDER BY salary DESC 
```

### **Quais são 100 funcionários mais bem pagos e suas posições?**

Com o intuito de retornar informações sobre os 100 funcionários com salários mais altos, declarou-se "SELECT TOP 100 e.first\_name, e.last\_name, e.position\_title, s.salary", ou seja, cada linha deveria conter as colunas "first\_name" e "last\_name" da tabela "employees", bem como as colunas "position\_title" e "salary" da tabela "salary". A cláusula "INNER JOIN salary s ON e.last\_name = s.last\_name" realizou a junção interna entre as tabelas "employees" e "salary". A junção é feita com base na coluna "last\_name", presente em ambas as tabelas. Por fim, indicou-se que os resultados da consulta deveriam ser ordenados com base na coluna "salary" em ordem decrescente, utilizando a cláusula "ORDER BY s.salary DESC".

```python
SELECT TOP 100 e.first_name, e.last_name, e.position_title, s.salary
FROM employees e
INNER JOIN salary s ON e.last_name = s.last_name
ORDER BY s.salary DESC
```
